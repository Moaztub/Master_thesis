# Base-scenario:
### 1) There are no fixed EV charging schedules. Owners charge as needed, regardless of electricity prices or hours of surplus solar, much like refuelling a petrol/diesel car.
### 2) There is no BESS to buffer surplus solar. Solar is used as needed; excess is curtailed. The remaining demand is met by a baseload biomass generator.
### 3) No demand-side flexibility is available from the flexible demand unit (flex_load). Cold-storage warehouse of this demand unit need to be operated for four consecutive hours per day; these hours are predefined and unaffected by electricity price signals.

```python
from demandlib.bdew import ElecSlp
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

e_slp = ElecSlp(year=2023)                     # general business
g0_annual = 12000

profile_g0 = e_slp.get_scaled_profiles({"g0": g0_annual})

demand_Gewerbe = profile_g0["g0"].resample("1h").sum()
demand_Gewerbe = demand_Gewerbe["2023-03-01":"2023-03-31"] / 1000
demand_Gewerbe= demand_Gewerbe.to_frame(name="Demand [MWh]")
#display(demand_Gewerbe.sum())

np.random.seed(42)

#  Generate 12 varied profiles
gewerbe_variants_profiles = []

for i in range(12):
    variation = np.random.uniform(0.8, 1.2, size=len(demand_Gewerbe))
    modified_values = demand_Gewerbe["Demand [MWh]"].values * variation
    
    df_variant = pd.DataFrame(modified_values, index=demand_Gewerbe.index, columns=[f"demand_Gewerbe_{i+1}_[MWh]"])
    
    gewerbe_variants_profiles.append(df_variant)

all_gewerbe_variants = pd.concat(gewerbe_variants_profiles, axis=1)

all_gewerbe_variants.iloc[:, 0].plot(figsize=(15, 4), title="demand Gewerbe Variant 1 (±10%)")
plt.xlabel("Time")
plt.ylabel("Electricity Demand [MWh]")
plt.grid(True)
plt.tight_layout()
plt.show()
tot_Gewerbe = all_gewerbe_variants.sum(axis=1).to_frame(name="tot_demand_Gewerbe_variants_[MWh]")
display(tot_Gewerbe)
#all_gewerbe_variants.head(5)
```

```python
from matplotlib.patches import Patch
start = "2023-03-01"
end   = "2023-03-31"
ev_name = "vw_id3_pro (2)"  # adjust as needed

# --- Prep data ---
charge_series = charging_profiles[ev_name].loc[start:end]  
solar_mw = solar_gen["power[W]"].loc[start:end].rename("Solar [MW]")

avail = (charge_series > 0).astype(int).rename("Availability (0/1)")

def contiguous_spans(binary_series, step="1h"):
    """Return list of (start, end) for 1-runs in a 0/1 Series sampled at `step`."""
    step = pd.Timedelta(step)
    on = binary_series[binary_series == 1]
    if on.empty:
        return []
    times = on.index.sort_values()
    starts, ends = [times[0]], []
    for prev, curr in zip(times[:-1], times[1:]):
        if curr - prev > step:
            ends.append(prev + step)
            starts.append(curr)
    ends.append(times[-1] + step)
    return list(zip(starts, ends))

charge_spans = contiguous_spans(avail, step="1h")

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(20, 8), sharex=True)


line_solar, = ax1.plot(solar_mw.index, solar_mw, color="orange", label="Solar [MW]", linewidth=1.5)

for sps, eps in charge_spans:
    ax1.axvspan(sps, eps, color="green", alpha=0.2, zorder=0)
charge_patch = Patch(facecolor="green", alpha=0.2, label="Charging window")

ax1.set_ylabel("Solar [MW]")
ax1.set_title(f"Solar & Charging Windows — {ev_name}  ({start} → {end})")
ax1.grid(True)
ax1.legend(handles=[line_solar, charge_patch], loc="upper right", frameon=True)

ax2.step(avail.index, avail, where="post", linewidth=1.5, label="Availability (0/1)")
for sps, eps in charge_spans:
    ax2.axvspan(sps, eps, color="limegreen", alpha=0.25, zorder=0)

ax2.set_ylim(-0.1, 1.1)
ax2.set_yticks([0, 1]); ax2.set_yticklabels(["Not avail.", "Avail."])
ax2.set_xlabel("Time")
ax2.set_title("EV charging availability")
ax2.grid(True, axis="x")
ax2.legend(handles=[Patch(facecolor="limegreen", alpha=0.25, label="Available window")],
           loc="upper right", frameon=True)

fig.tight_layout()
plt.show()
```

# Cumulative Supply vs. Total Demand

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ----- Window -----
start_date = "2023-03-01"
end_date   = "2023-03-10 23:00"

# ----- Supply side (MW) -----
biomass = net.generators_t.p["Biomass"].squeeze().loc[start_date:end_date]
solar   = net.generators_t.p["solar"].squeeze().loc[start_date:end_date]
grid    = net.generators_t.p["grid_generator"].squeeze().loc[start_date:end_date]
demand  = Tot_demand.squeeze().loc[start_date:end_date]               # MW (total demand power)
rated_solar = solar_gen.squeeze().loc[start_date:end_date]            # MW (installed/rated solar output)
mask = rated_solar > demand

# Align all to a common index
common_index = solar.index.intersection(biomass.index)\
                           .intersection(grid.index)\
                           .intersection(demand.index)\
                           .intersection(rated_solar.index)
biomass = biomass.loc[common_index]
solar   = solar.loc[common_index]
grid    = grid.loc[common_index]
demand  = demand.loc[common_index]
rated_solar = rated_solar.loc[common_index]
mask    = mask.reindex(common_index, fill_value=False)

# Cumulative for stacked fills
solar_cum   = solar
biomass_cum = solar_cum + biomass
grid_cum    = biomass_cum + grid

# ----- Demand breakdown (MWh) -----
# If `snapshots` isn't defined, take it from the network; then reindex to common_index
if "snapshots" not in globals():
    snapshots = net.snapshots

df_Tot_demand = pd.DataFrame({
    "Households+PV+storage [MWh]":                 tot_Houshold_pv_storage["tot_demand_Houshold_pv_storage_variants_[MWh]"],
    "Households [MWh]":                            tot_Houshold["tot_demand_Houshold_variants_[MWh]"],
    "Continuous business [MWh]":                   tot_flex_load["tot_demand_flex_load_variants_[MWh]"],
    "Business on weekdays 8 a.m. - 6 p.m. [MWh]":  tot_Landswirtschaftsbetriebe["tot_demand_Landschaftsbetriebe_variants_[MWh]"] 
                                                   if "tot_demand_Landschaftsbetriebe_variants_[MWh]" in tot_Landswirtschaftsbetriebe
                                                   else tot_Landswirtschaftsbetriebe["tot_demand_Landswirtschaftsbetriebe_variants_[MWh]"],
    "general business [MWh]":                      tot_Gewerbe["tot_demand_Gewerbe_variants_[MWh]"],
    "EVs charging [MWh]":                          Tot_EVs_charging["EVs demand [MWh]"],
}).reindex(snapshots).fillna(0).astype(float)

# focus on same time window as top chart
df_Tot_demand_window = df_Tot_demand.reindex(common_index).fillna(0)

# ----- Plot (two stacked subplots, shared x) -----
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(16, 9), sharex=True)

# === Top: cumulative supply vs demand (MW) ===
ax1.fill_between(common_index, 0,           solar,        label="Solar",          alpha=0.7, color="orange")
ax1.fill_between(common_index, solar,       biomass_cum,  label="Biomass",        alpha=0.7, color="green")
ax1.fill_between(common_index, biomass_cum, grid_cum,     label="Grid Generator", alpha=0.7, color="brown")

# Solar curtailment: area where rated_solar > demand
ax1.fill_between(common_index, demand, rated_solar, where=mask, interpolate=True,
                 label="Solar Curtailment", alpha=0.7, color="grey")

ax1.plot(common_index, demand,      label="Total Demand", color="black", linewidth=2)
ax1.plot(common_index, rated_solar, label="Rated Solar",  color="red",   linewidth=1)

ax1.set_ylabel("Power [MW]")
ax1.set_title("Cumulative Supply vs. Total Demand")
ax1.grid(True, axis="both", linestyle="-", linewidth=0.5, alpha=0.6)
ax1.legend(loc="upper left", ncol=3, frameon=True)

# === Bottom: demand breakdown (MWh) ===
# stacked area: columns order controls stacking
cols = list(df_Tot_demand_window.columns)
ax2.stackplot(df_Tot_demand_window.index,
              *[df_Tot_demand_window[c] for c in cols],
              labels=cols, alpha=0.8)

ax2.set_ylabel("Energy [MWh]")
ax2.set_xlabel("Time")
ax2.set_title("Demand Breakdown (Energy per Snapshot)")
ax2.grid(True, axis="x", linestyle="-", linewidth=0.5, alpha=0.6)
ax2.legend(loc="upper left", ncol=2, frameon=True)

fig.tight_layout()
plt.show()
```
